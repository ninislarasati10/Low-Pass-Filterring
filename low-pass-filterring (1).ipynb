```python
# memanggil modul yang diperlukan
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl

#bgr
img = cv2.imread('../input/wanita/Screenshot (1194).png')

#rgb
photo = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# tampilkan gambar awal tanpa filter
plt.imshow(img)
plt.show()

# membuat filter: matriks berukuran 5 x 5
kernel = np.ones((5,5),np.float32)/25
print(kernel)

# lakukan filtering
photo_filter = cv2.filter2D(img,-1,kernel)

plt.imshow(photo_filter)
```

```python
# perbesar ukuran hasil plotting
plt.rcParams["figure.figsize"] = (15,15)

# plot pertama, gambar asli
plt.subplot(121),plt.imshow(photo),plt.title('Original')
plt.xticks([]), plt.yticks([])

# kedua, hasil filter
plt.subplot(122),plt.imshow(photo_filter),plt.title('Averaging')
plt.xticks([]), plt.yticks([])

# Plot!
plt.show()
```

```python
photo_blur = cv2.blur(img,(5,5))

plt.imshow(photo_blur)
```

```python
# ini adalah cara lain untuk membuat sebuah kernel, 
# yaitu dengan menggunakan np.matrix
# kali ini, ukuran matriksnya 3 x 3
kernel = np.matrix([
          [1, 1, 1],
          [1, 2, 1],
          [1, 1, 1]         
          ])/25
print(kernel)

# buat lagi filteringnya
photo_filter = cv2.filter2D(img,-1,kernel)

# tampilkan
plt.imshow(photo_filter)
```
